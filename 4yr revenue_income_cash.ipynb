```python
from yahoofinancials import YahooFinancials
import pandas as pd
import matplotlib.pyplot as plt
import requests
import bs4 as bs
```

```python
resp = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = bs.BeautifulSoup(resp.text,"lxml")
table = soup.find('table',{'class':'wikitable sortable'})
tickers = []
for row in table.findAll('tr')[1:]:
    ticker = row.findAll('td')[0].text.replace('.','-')
    ticker = ticker[:-1]
    tickers.append(ticker)
```

```python
stocks = ['AAPL', 'MSFT', 'INTC', 'FRC', 'TSLA','EW']
```

```python
yahoo_financials_list = YahooFinancials(tickers)
df = yahoo_financials_list.get_financial_stmts('annual', ['income', 'cash'])
```

```python
count = 1
count2 = 1
for eachstock in tickers:
    
    try:

        df1 = pd.DataFrame(df['incomeStatementHistory'][eachstock][3])
        df2 = pd.DataFrame(df['incomeStatementHistory'][eachstock][2])
        df3 = pd.DataFrame(df['incomeStatementHistory'][eachstock][1])
        df4 = pd.DataFrame(df['incomeStatementHistory'][eachstock][0])

        df5 = pd.DataFrame(df['cashflowStatementHistory'][eachstock][3])
        df6 = pd.DataFrame(df['cashflowStatementHistory'][eachstock][2])
        df7 = pd.DataFrame(df['cashflowStatementHistory'][eachstock][1])
        df8 = pd.DataFrame(df['cashflowStatementHistory'][eachstock][0])

        year1=df1.columns[0]
        year2=df2.columns[0]
        year3=df3.columns[0]
        year4=df4.columns[0]

        tr1=df1[year1].loc['totalRevenue']
        tr2=df2[year2].loc['totalRevenue']
        tr3=df3[year3].loc['totalRevenue']
        tr4=df4[year4].loc['totalRevenue']

        oi1=df1[year1].loc['operatingIncome']
        oi2=df2[year2].loc['operatingIncome']
        oi3=df3[year3].loc['operatingIncome']
        oi4=df4[year4].loc['operatingIncome']

        cfo1=df5[year1].loc['totalCashFromOperatingActivities']
        cfo2=df6[year2].loc['totalCashFromOperatingActivities']
        cfo3=df7[year3].loc['totalCashFromOperatingActivities']
        cfo4=df8[year4].loc['totalCashFromOperatingActivities']
                                                        
        if tr2 > tr1 and tr3 > tr2 and tr4 > tr3 and oi2 > oi1 and oi3 > oi2 and oi4 > oi3 and cfo1 > oi1 and cfo2 > oi2 and cfo3 > oi3 and cfo4 > oi4:
            print(count, count2, " TR & OI increase 4 yrs in a row : " + eachstock)
            count = count+1
            count2 = count2 + 1
        else:
            print(count, " Not this one : " + eachstock)
            count = count+1
    except:
        print(count, " Error : " + eachstock)
        count = count+1
        pass
```
